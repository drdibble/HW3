```python
import gzip
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split

from sklearn.metrics import plot_confusion_matrix, f1_score, accuracy_score
from sklearn.neighbors import KNeighborsClassifier


def load_fashion_mnist():
    """
    Loads Fashion MNIST dataset.
    
    Adapted from: https://github.com/zalandoresearch/fashion-mnist/blob/master/utils/mnist_reader.py
    """
    TRAIN_IMAGES = 'data/train-images-idx3-ubyte.gz'
    TRAIN_LABELS = 'data/train-labels-idx1-ubyte.gz'    
    TEST_IMAGES = 'data/t10k-images-idx3-ubyte.gz'
    TEST_LABELS = 'data/t10k-labels-idx1-ubyte.gz'

    with gzip.open(TRAIN_LABELS, 'rb') as tr_labels_file, gzip.open(TEST_LABELS, 'rb') as ts_labels_file:
        train_labels = np.frombuffer(tr_labels_file.read(), dtype=np.uint8, offset=8)
        test_labels = np.frombuffer(ts_labels_file.read(), dtype=np.uint8, offset=8)

    with gzip.open(TRAIN_IMAGES, 'rb') as tr_images_file, gzip.open(TEST_IMAGES, 'rb') as ts_images_file:
        train_images = np.frombuffer(tr_images_file.read(), dtype=np.uint8, offset=16).reshape(len(train_labels), 784)
        test_images = np.frombuffer(ts_images_file.read(), dtype=np.uint8, offset=16).reshape(len(test_labels), 784)

    return train_images, train_labels, test_images, test_labels
```

```python
def pretty_print(image_example):
    """ Pretty prints a Fashion MNIST example.

    Parameters:
        image_example: a 1x784 numpy array corresponding to the features of
                       a single image.

    Returns:
        None.
    """
    print(np.array_str(image_example, precision=1, max_line_width=116))
```

```python
def usage_example():
    """ Example of how to load and parse Fashion MNIST data. """
    
    train_images, train_labels, test_images, test_labels = load_fashion_mnist()

    # train_images is a 60,000 x 784 numpy matrix. There are 60k
    # rows in the matrix, each row corresponding to a single example.
    # There are 784 columns, each corresponding to the value of a
    # single pixel in the 28x28 image after it has been "flattened".
    print("Dimensions of training set feature matrix:", train_images.shape)

    # The labels for each example are maintained separately in train_labels.
    # This is a 60,000 x 1 numpy matrix, where each element is the label
    # for the corresponding training example.
    print("Dimensions of training set label matrix:", train_labels.shape)

    # Example of how to access a individual training example (in this case,
    # we pick an example at a random index). We could use print to output the
    # raw pixel values to the screen, but pretty_print formats the data in 
    # a nicer way: if you squint, you may be able to make out the contours of
    # the fashion article in the matrix data.
    EXAMPLE_INDEX = np.random.randint(60000)
    print("Features of training example at index {}:\n".format(EXAMPLE_INDEX))
    pretty_print(train_images[EXAMPLE_INDEX])

    # And here's the label that goes with that training example
    print("\nLabel of training example at index {}:".format(EXAMPLE_INDEX), train_labels[EXAMPLE_INDEX], '\n')

    # Finally, let's visualize the example we've picked as a 28x28 image
    plt.figure()
    plt.imshow(train_images[EXAMPLE_INDEX].reshape(28, 28))
    plt.show()
    
    # The test_images/test_labels are organized in the same way, but only contain 10k
    # examples. Don't touch this data until your model is frozen! Perform all
    # cross-validation, model selection, hyperparameter tuning etc. on the 60k
    # training set. Use the test set simply for reporting performance.
```

```python
'''
Estimates logit model
Parameter sets should be scaled before being passed into this function
'''
def logit_model(train_images, train_labels, test_images, test_labels):
    
    print('Logit model with newton-cg solver...')
    model = LogisticRegression(penalty='l2', solver='newton-cg', multi_class='multinomial', max_iter=500)
    model.fit(train_images, train_labels)
    
    print('Logit model with sag solver...')
    model2 = LogisticRegression(penalty='l2', solver='sag', multi_class='multinomial', max_iter=500)
    model2.fit(train_images, train_labels)
    
    print('Performance:')
    newton_preds = model.predict(test_images)
    newton_f_score = f1_score(test_labels, newton_preds, average='micro')
    sag_preds = model2.predict(test_images)
    sag_f_score = f1_score(test_labels, sag_preds, average='micro')
    print('newton-cg F1: %f' % newton_f_score)
    print('sag F1: %f' % sag_f_score)
    
    print('Done!')
    if newton_f_score > sag_f_score:
        print('Returning newton-cg model')
        return model
    else:
        print('Returning sag model')
        return model2
```

```python
'''
Estimates knn models for several values of k, then returns the best performing model (by F1 score)
Parameter sets should be scaled before being passed into this function
'''
def k_NN_model_micro(train_images, train_labels, test_images, test_labels):
    
    print('Estimating knn model:')
    
    k_arr = np.array([])
    f1_arr = np.array([])
    acc_arr = np.array([])
    model_arr = np.array([])
    
    for k in range(1,35,2):
        print('k = %d' % k)
        classifier = KNeighborsClassifier(n_neighbors=k)
        classifier.fit(train_images, train_labels)
        model_arr = np.append(model_arr, classifier)
        y_preds = classifier.predict(test_images)
        f_score = f1_score(test_labels, y_preds, average='micro')
        k_arr = np.append(k_arr, k)
        f1_arr = np.append(f1_arr, f_score)
        acc_score = accuracy_score(test_labels, y_preds)
        acc_arr = np.append(acc_arr, acc_score)
    
    fig = plt.figure(1)
    plt.scatter(k_arr, f1_arr)
    plt.title('k vs F1 Score (knn model)')
    plt.xlabel('k') 
    plt.ylabel('F1 Score')
    #ax2.scatter(k_arr, acc_arr)   
    #ax2.set_title('k vs Accuracy Score - Micro')
    #ax2.set_xlabel('k') 
    #ax2.set_ylabel('Accuracy Score')
    plt.savefig('imgs/k_selection.png', bbox_inches='tight')
    plt.show()
    
    best_k_index = np.argmax(f1_arr)
    return k_arr[best_k_index], model_arr[best_k_index]
```

```python
def main():
    #load data
    #evaluation set should only be used for model evaluation at the end of training
    train_images, train_labels, eval_images, eval_labels = load_fashion_mnist()
    
    #80/20 train/test split to test hyperparams
    X_train, X_test, y_train, y_test = train_test_split(train_images, train_labels, test_size=0.2, random_state=2)
    
    #scale data
    min_max_scalar = preprocessing.MinMaxScaler()
    X_train_scaled = min_max_scalar.fit_transform(X_train)
    min_max_scalar = preprocessing.MinMaxScaler()
    X_test_scaled = min_max_scalar.fit_transform(X_test)
    min_max_scalar = preprocessing.MinMaxScaler()
    eval_images_scaled = min_max_scalar.fit_transform(eval_images)
    
    #find best logit model
    logit = logit_model(X_train_scaled, y_train, X_test_scaled, y_test)
    
    #find best knn model
    k, knn = k_NN_model_micro(X_train_scaled, y_train, X_test_scaled, y_test)
    
    #calculate accuracy and F1 score for each model w/ evaluation set
    
    #logit model
    logit_preds = logit.predict(eval_images_scaled)
    logit_f_score = f1_score(eval_labels, logit_preds, average='micro')
    logit_acc_score = accuracy_score(eval_labels, logit_preds)
    
    #knn model
    knn_preds = knn.predict(eval_images_scaled)
    knn_f_score = f1_score(eval_labels, knn_preds, average='micro')
    knn_acc_score = accuracy_score(eval_labels, knn_preds)
    
    print('Logit: F1 = %f, accuracy = %f' % (logit_f_score, logit_acc_score))
    print('KNN: k = %d, F1 = %f, accuracy = %f' % (k, knn_f_score, knn_acc_score))
    
    #confusion matrices
    plot_confusion_matrix(logit, eval_images_scaled, eval_labels)
    plt.title('Confusion Matrix for Logit Model')
    plt.savefig('imgs/logit_confusion.png')
    plt.show()
    
    plot_confusion_matrix(knn, eval_images_scaled, eval_labels)
    plt.title('Confusion Matrix for knn Model')
    plt.savefig('imgs/knn_confusion.png')
    plt.show()
```

```python
main()
```
